```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
data
```

```python
data.columns
```

```python
plt.figure(figsize=(10, 5))
plt.bar(data['year'], data['tot_hhs'], color='blue', label='household')
plt.xlabel('year')
plt.ylabel('tot_hhs')
plt.title('yearly house')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(data['year'], data['income'], marker='o', linestyle='-', color='blue', label='Income')
plt.plot(data['year'], data['expenditure'], marker='o', linestyle='-', color='red', label='Expenditure')

plt.xlabel('Age')
plt.ylabel('amount')
plt.title('income and expenditure by age group')
plt.legend()

plt.show()
```

```python
identifier_column = 'year'  # Adjust this if your identifier column has a different name

for column in data.columns:
    if column != identifier_column:
        plt.figure(figsize=(10, 6))
        plt.plot(data[identifier_column], data[column], marker='o', label=column)
        plt.xlabel(identifier_column)
        plt.ylabel(column)
        plt.title(f'Line Chart for {column}')
        plt.legend()
        plt.show()
```
