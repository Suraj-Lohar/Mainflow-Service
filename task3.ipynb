```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
data
```

```python
df=pd.DataFrame(data)

df.columns
```

```python
plt.scatter(df['year'],df['tot_hhs'],color="red",label="household")
plt.title (" Scatter Plot")

plt.xlabel ( 'year')
plt.ylabel('total house' )
plt.colorbar ()
plt.show ()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['year'], df['tot_hhs'], color='blue', label='household')
plt.xlabel('year')
plt.ylabel('tot_hhs')
plt.title('yearly house')
plt.legend()
plt.show()
```

```python
# Creating a line chart for 'income' and 'expenditure'
plt.figure(figsize=(8,4))
plt.plot(df['year'], df['income'], marker='o', linestyle='-', color='green', label='Income')
plt.plot(df['year'], df['expenditure'], marker='x', linestyle='--', color='red', label='Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Income vs Expenditure Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
identifier_column = 'year'  

for column in df.columns:
    if column != identifier_column:
        plt.figure(figsize=(10, 6))
        plt.plot(df[identifier_column], df[column], marker='o', label=column)
        plt.xlabel(identifier_column)
        plt.ylabel(column)
        plt.title(f'Line Chart for {column}')
        plt.legend()
        plt.show()
```
